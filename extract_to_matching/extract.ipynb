```python
# !pip install pdfminer
# !pip install PyPDF2
import io
from pdfminer.converter import TextConverter
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfpage import PDFPage
import PyPDF2
import os
import math

def get_pdf_text(inputFileName,a):
    text_list = []
    with open(inputFileName, 'rb') as fh:
        for page in PDFPage.get_pages(fh, caching=True, check_extractable=True):
            resource_manager = PDFResourceManager()
            fake_file_handle = io.StringIO()
            converter = TextConverter(resource_manager, fake_file_handle)
            page_interpreter = PDFPageInterpreter(resource_manager, converter)
            page_interpreter.process_page(page)
            s = fake_file_handle.getvalue()
            s = s.split("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
            converter.close()
            fake_file_handle.close()
            try:
                text_list.append(get_name(s[a]))
            except:
                continue
    return text_list

def get_pdf_roll_no(data):
    import re
    roll_no = re.findall(r'\b\d{10}\b', data)
    return roll_no

def get_name(data):
    name = []
    roll_number = []
    temp = []
    text = data.split(' ')
    n = len(text)
    c = 0
    while c<n:
        if text[c] != "":
            try:
                p = text[c].replace('.',"")
                p = p.strip()
                p = int(p)
                if math.floor(math.log10(p)+1) == 10:
                    roll_number.append(p)
            except:
                try:
                    p = text[c+1].replace('.',"")
                    p = int(p)
                    name.append(text[c])
                except:
                    try:
                        p = text[c+2].replace('.',"")
                        p = int(p)
                        name.append(text[c]+" "+text[c+1].strip)
                        c += 1
                    except:
                        name.append(text[c] + " " + text[c+1] + " "+ text[c+2])
                        c += 2
        c+=1
    for x in name:
        if x.find('0') != -1:
            name.remove(x)
        
    return name
    
def get_pdf_data(inputFileName,a):
    text = get_pdf_text(inputFileName,a)
    temp = []
    for t in text:
        for x in t:
            x = x.strip()
            if x not in l and len(x) > 4 and x.find("\\") == -1:
                temp.append(x)
    return temp
```
